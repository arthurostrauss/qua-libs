```python
from qiskit import transpile
from qiskit.circuit.library import QFT
```

```python
qc = QFT(5)

transpiled_qc = transpile(qc, basis_gates=['x','sx','rz','cz'], optimization_level=3, coupling_map=[[0,1],[1,2],[2,3],[3,4],[4,5]])
```

```python
transpiled_qc.draw('mpl')
```

```python
from qiskit.qasm3 import dumps
qasm = dumps(transpiled_qc, includes=(),basis_gates=['x','sx','rz','cz'])

print(qasm.replace("$1","q[0]").replace("$2","q[1]").replace("$3","q[2]").replace("$4","q[3]").replace("$5","q[4]"))
```
